```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("15_sb.xlsx")
data.columns = data.columns.str.lower()
```

```python
data
```

## ¬øCu√°l es el Municipio que m√°s ha votado y por qu√© partido?

```python
M = data.groupby("municipio")[["vottot"]].sum()
M.sort_values(by='vottot', ascending = False).head(4)
```

El municipio que m√°s ha votado es Ecatepec con 13,926,606 votos seguido de Nezahualcoy√≥tl, Naucalpan y Tlanepantla.

```python
M.sort_values(by = 'vottot',ascending = False).head(4).plot.bar(rot=0,figsize = (10,5),color = 'purple', 
                                                                  title = 'Votos Totales por Municipio')
plt.yticks(color="black")
plt.xticks(color="black", fontsize = 10 )
plt.title('Votos por Municipio', fontsize = 12)
plt.xlabel('Municipio', fontsize = 12)
plt.ylabel('Votos en escala de 10 millones')
```

```python
M2 = data.groupby("municipio")[["pri",'pan','prd','pt','pvem','mc','na','morena']].sum()
M2 = M2.sort_values(by='pri', ascending = False)
M2.head()
```

El partido por el que m√°s ha votado Ecatepec es el PRI

```python
M2.sort_values(by='pri', ascending = False).head(10).plot.barh(figsize=(8,15), title='Votos Totales por Municipio')
plt.yticks(color="black")
plt.xticks(color="black")
plt.title('Votos Totales por Municipio', fontsize = 15)
plt.xlabel('Votos', fontsize = 15)
plt.ylabel('Municipio',fontsize = 15)
```

## ¬øQu√© partido tiene mayor control del Estado?

```python
M3 = data.groupby("a√±o")[["pri",'pan','prd','morena','pvem','mc','na','pt']].sum()
M3
```

El partido que tiene mayor control del estado de M√©xico es el PRI con 8 elecciones ganadas de 11 desde 1997 hasta 2017
